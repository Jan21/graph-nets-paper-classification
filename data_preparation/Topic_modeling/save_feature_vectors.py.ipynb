```python
import numpy as np
import pickle
from collections import Counter

with open('id_and_topic.pickle', 'rb') as f:
    id2topic = pickle.load(f)
#id2topic_vector = {k:np.array([tup[1] for tup in tups]) for k, tups in id2topic.items()}

counter =  Counter()
repaired_id2topic = {}
for k,v in id2topic.items():
    if len(v) == 24:
        repaired_id2topic[k] = np.array([val[1] for val in v])
        continue
    pos,vals = list(zip(*v))
    pos_val_dict = {}
    for p in range(len(pos)):
        pos_val_dict[pos[p]] = vals[p]
    
    arr_of_vals = []
    for i in range(24):
        if i in pos:
            arr_of_vals.append(pos_val_dict[i])
        else:
            arr_of_vals.append(0)
    if len(arr_of_vals)!= 24:
        print('something went wrong')
    repaired_id2topic[k] = np.array(arr_of_vals)
        
for k,v in repaired_id2topic.items():
    if len(v)!=24:
        print('error')
```

## save lables

```python
import json
id2label_dict = {}
venue_counter = Counter()
with open('../data_connected_graph.jsonl') as f:
    a = []
    for line in f.readlines():
        dic = json.loads(line)
        id2label_dict[dic['id']] = dic['venue']
        venue_counter[dic['venue']] += 1
print(len(id2label_dict.items()))
print(venue_counter)
```

```python
venue2num = {'ACL':0, 'ML':1, 'AI': 2}
arr_of_features = []
arr_of_labels = []
for i in range(len(line2id)):
    arr_of_features.append(repaired_id2topic[line2id[i]])
    num_label = venue2num[id2label_dict[line2id[i]]]
    arr_of_labels.append(num_label)

np_arr_of_features = np.vstack(arr_of_features)
np_arr_of_labels = np.array(arr_of_labels, dtype=int)
print(len(arr_of_features),len(arr_of_labels))
with open('input_featuers_abstracts.pkl','wb') as f:
    pickle.dump(np_arr_of_features,f,pickle.HIGHEST_PROTOCOL)
with open('lables.pkl','wb') as f:
    pickle.dump(np_arr_of_labels,f,pickle.HIGHEST_PROTOCOL)
```
